#### Since the columns "clock_speed", "m_dep" and "n_cores" have very low correlations to the target variable "price_range", we would drop them

```python
df = df.drop(columns= ["clock_speed", "m_dep", "n_cores"])
df
```

#### Scaling inputs for modelling

###### Defining the x and y

```python
x = df.drop(columns="price_range")
y = df['price_range']
```

###### Splitting into train, val and test data for both models

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

###### scaling

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_val_scaled = scaler.transform(x_val)
x_test_scaled = scaler.transform(x_test)
```

##### Saving data in npz format for tensorflow

```python
np.savez('Mobile_price_train', inputs=x_train_scaled, targets=y_train)
np.savez('Mobile_price_validation', inputs=x_val_scaled, targets=y_val)
np.savez('Mobile_price_test', inputs=x_test_scaled, targets=y_test)
```

##### Saving the Scaler for consistency

```python
import pickle
```

```python
with open('scaler', 'wb') as f:
    pickle.dump(scaler, f)
```

##### Building a Logistic Regression with sklearn

```python
log_reg = LogisticRegression(C=20)
```

```python
log_reg.fit(x_train_scaled, y_train)
```

```python
y_pred = log_reg.predict(x_val_scaled)
accuracy = accuracy_score(y_val, y_pred)
print("Validation Accuracy: ", accuracy)
```

```python
log_reg_train = log_reg.score(x_train_scaled, y_train)
print(f"Train accuracy: {log_reg_train}")
```

```python
y_pred_test = log_reg.predict(x_test_scaled)
test_accuracy = accuracy_score(y_test,y_pred_test)
print("Testing Accuracy: ", test_accuracy)
```

##### Testing on a seperate dataset

```python
test_data = pd.read_csv('test.csv')
test_data = test_data.drop(columns=["id", "clock_speed", "m_dep", "n_cores"])
```

```python
scaled_test_data = scaler.transform(test_data)
```

```python
y_pred_new_test = log_reg.predict(scaled_test_data)
new_test_accuracy = log_reg.score(scaled_test_data, y_pred_new_test)
print("Testing Accuracy: ", test_accuracy)
```
